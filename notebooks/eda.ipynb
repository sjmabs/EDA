```python
# load the dataset
import pandas as pd
df = pd.read_csv("./data/UK-HPI-full-file-2024-RAW.csv")
df
```

```python
# convert date column to datatime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df
```

```python
# checking for uniqueness of region names
regions = df["RegionName"].unique()
regions
```

```python
# extract the year
df["Year"] = df["Date"].dt.year

# group by year and calc average property price
yearly_avg_price = df.groupby("Year")["AveragePrice"].mean().reset_index()
yearly_avg_price
```

```python
# line chart showing average price over time
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(yearly_avg_price["Year"], yearly_avg_price["AveragePrice"], marker='o', linestyle='-', color='b')

plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title('Average House Price Trends (UK)')
plt.grid(True)
plt.show()

```

```python
# average flat price removing years that have no flat price data
yearly_avg_flat_price = df.groupby("Year")["FlatPrice"].mean().dropna().reset_index()
yearly_avg_flat_price
```

```python
plt.figure(figsize=(10, 5))
plt.plot(yearly_avg_flat_price["Year"], yearly_avg_flat_price["FlatPrice"], marker='o', linestyle='-', color='b')

plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title('Average Flat Price Trends (UK)')
plt.grid(True)
plt.show()
```

```python
average_FTB_price = df.groupby("Year")["FTBPrice"].mean().dropna().reset_index()
average_FTB_price
```

```python
plt.figure(figsize=(10, 5))
plt.plot(average_FTB_price["Year"], average_FTB_price["FTBPrice"], marker='o', linestyle='-', color='b')

plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title('Average First Time Buyer Price Trends (UK)')
plt.grid(True)
plt.show()
```

```python
# Regional variations - comparing prices across different countries

yearly_avg_price_country = df.groupby(["Year", "RegionName"])["AveragePrice"].mean().reset_index()
yearly_avg_price_country = yearly_avg_price_country[yearly_avg_price_country["RegionName"].isin(["England", "Northern Ireland", "Scotland", "Wales"])]
# Pivot the data so each region has its own column
yearly_avg_price_country = yearly_avg_price_country.pivot(index='Year', columns='RegionName', values='AveragePrice')

yearly_avg_price_country
```

```python
plt.figure(figsize=(10, 5))
for column in yearly_avg_price_country.columns:
    plt.plot(yearly_avg_price_country.index, yearly_avg_price_country[column], marker='o', linestyle='-', label=column)

plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title('Average House Price Trends by Country (UK)')
plt.legend(title="Region", loc="upper left", fontsize=10, frameon=True)  # Add legend with title
plt.grid(True)
plt.show()
```
