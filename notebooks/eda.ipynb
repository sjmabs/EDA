```python
# load the dataset
import pandas as pd
df = pd.read_csv("./data/UK-HPI-full-file-2024-RAW.csv")
df
```

```python
# convert date column to datatime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df
```

```python
# checking for uniqueness of region names
regions = df["RegionName"].unique()
regions
```

```python
# extract the year
df["Year"] = df["Date"].dt.year

# group by year and calc average property price
yearly_avg_price = df.groupby("Year")["AveragePrice"].mean().reset_index()
yearly_avg_price
```

```python
# average flat price removing years that have no flat price data
yearly_avg_flat_price = df.groupby("Year")["FlatPrice"].mean().dropna().reset_index()
yearly_avg_flat_price
```

```python
average_FTB_price = df.groupby("Year")["FTBPrice"].mean().dropna().reset_index()
average_FTB_price
```

```python
# Regional variations - comparing prices across different countries

yearly_avg_price_country = df.groupby(["Year", "RegionName"])["AveragePrice"].mean().reset_index()
yearly_avg_price_country = yearly_avg_price_country[yearly_avg_price_country["RegionName"].isin(["England", "Northern Ireland", "Scotland", "Wales"])]
yearly_avg_price_country
```
